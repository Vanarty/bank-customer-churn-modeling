#### Обработка пропусков

Так как объем выборки небольшой (всего 10000 объектов) удаление объектов с пропущенным значением в столбце `['tenure']` может привести к потере почти 10% данных с теми признаками, которые потенциально могут быть очень важны для обучения моделей. Лучшей стратегией в данном случае будет восстановление пропущенных значений признака по какому-либо другому столбцу/столбцам, которые: 
1) имеют связь с этим признаком (корреляция, логическая обоснованность и т.д.); 

2) размерность значений по одному из таких столбцов должна быть больше, или хотя бы равна размерности значений восстанавливаемого признака.

```python
df.describe()
```

1. Наилучшим признаком по которому можно восстановить данные в столбце `['tenure']` является возраст клиента (`['age']`). Так как количество лет, когда человек является клиентом банка может опосредовано говорить нам о его примерном возрасте. Условно говоря, чем старше человек, тем выше шанс его долгосрочного сотрудничества с банком (например врядли молодые люди 18 лет имеют более 1 года сотрудничества с банком в силу возраста). Так же неплохим решением будет использование признака `['geography']` для ещё большего разбиения выборки клиентов по группам для подсчета медианы по признаку `['tenure']`, так как география клиентов может косвенно влиять на их поведение и ментальность. 
2. Значений признака `['age']` явно больше чем у признака `['tenure']`, что дает возможность заполнить пропуски всеми возможными значениями диапазон которых от 0 до 10. 

Конечно, хоть возраст клиента (признак `['age']`) и может как-то коррелировать с признаком `['tenure']`, но связь эту нельзя назвать очевидной или значимой. Тем не менее для сохранения почти 10% данных с теми признаками, которые могут быть очень важны для обучения и правильного предсказания моделей восстановление с использованием медианы значений `['tenure']` по предварительно сгруппированным объектам по признаку `['age']` и `['geography']` является хорошей стратегией.

```python
df['tenure'] = df['tenure'].fillna(
    df[['age', 'geography', 'tenure']]
    .groupby(['age', 'geography'])['tenure']
    .transform('median')
    .round()
).astype('int8')

# проверим осталичь ли пропуски
df['tenure'].isna().sum()
```

### Разбивка данных на выборки

Подготовим два датафрейма для обучения моделей (**features** и **target**).

```python
# для обучения моделей оставим только необходимые признаки
features = df.drop(['row_number', 'customer_id', 'surname', 'exited'], axis=1)
target = df['exited']

# выведем размеры features и target
display(features.shape, target.shape)
```

В связи с тем, что в исходных данных отдельной тестовой выборки нет, данные необходимо разбить на обучающую, валидационную и тестовую выборку в соотношении 3:1:1 (60%, 20%, 20%).

```python
rand_st=12345

# разбивка на обучающую и обобщенную (валидационную + тестовую) выборки
features_train, features_valid_test, target_train, target_valid_test = train_test_split(
    features, target, test_size=0.4, stratify = target, random_state=rand_st
    )
# разбивка на валидационную и тестовую выборки
features_valid, features_test, target_valid, target_test = train_test_split(
    features_valid_test, target_valid_test, test_size=0.5, stratify = target_valid_test, random_state=rand_st
    )
display('Размер обучающей выборки:', features_train.shape, target_train.shape)
display('Размер валидационной выборки:', features_valid.shape, target_valid.shape)
display('Размер тестовой выборки:', features_test.shape, target_test.shape)
```

### Масштабирование численных признаков

Для таких моделей как логистическая регрессия важно, чтобы признаки имели значения с одним порядком (стандартизированы или нормализованы). Для этого применим метод масштабирования **MinMaxScaler** численных признаков `['credit_score']`, `['age']`, `['balance']`, `['num_of_products']`, `['estimated_salary']` и `['tenure']`.

```python
# составим список столбцов, которые необходимо масштабировать
cols_for_scale = ['credit_score', 'age', 'num_of_products', 'tenure', 'estimated_salary', 'balance']
# проведем масштабирвоание с применением MinMaxScaler
scaler = MinMaxScaler()
# настроим его на обучающих данных
scaler.fit(features_train[cols_for_scale])
# подготовим отдельные датафреймы с масштабированными признаками 
features_train_scale = features_train.copy()
features_valid_scale = features_valid.copy()
features_test_scale = features_test.copy()
# преобразуем обучающую, валидационную и тестовую выборки функцией transform()
features_train_scale[cols_for_scale] = scaler.transform(features_train[cols_for_scale])
features_valid_scale[cols_for_scale] = scaler.transform(features_valid[cols_for_scale])
features_test_scale[cols_for_scale] = scaler.transform(features_test[cols_for_scale])
# выведем для контроля данные обучающей выборки с отмасштабированными признаками
features_train_scale.sample(5)
```

Для модели подобранной на данной метрике (будем считать её лучшей) будет осуществлен подбор оптимального порога классификации путём его перебора и пересчета метрик `F1-мера`, `Accuracy` и `ROC-AUC` на валидационной выборке, поэтому перебор и поиск будет выполнен в ручную в циклах (без использования *GridSearchCV* например). #### Модель решающего дерева (DecisionTreeClassifier)

Для начала обучим **модель решающего дерева** с наилучшими значениями гиперпараметров **max_depth** и **max_features**. Для подбора гиперпараметров с кросс-валидацией будет использован **GridSearchCV**. 

```python
%%time 

# инициализируем перемнную для лучшей оценки f1-меры 
best_f1 = 0 
                      
# подберем лучший гиперпараметр max_depth в цикле 
for max_feature_ in range(1,12):
    for depth_ in range(1,12):
        model_tree = DecisionTreeClassifier(max_depth=depth_, max_features=max_feature_, random_state=rand_st)
        model_tree.fit(features_train_scale, target_train)
        predictions_valid = model_tree.predict(features_valid_scale)
      
        f1 = f1_score(target_valid, predictions_valid)
        
        # запишем в переменные гиперпараметры лучшей модели и саму модель
        if f1 > best_f1:
            best_f1 = f1
            max_feature = max_feature_
            depth = depth_
            best_model_tree = model_tree
        
# запишем в переменные и итоговые списки метрики лучшей модели 
predictions_valid = best_model_tree.predict(features_valid_scale)
accur = accuracy_score(target_valid, predictions_valid)
# для метрики ROC-AUC подсчитаем вероятности отнесения к классу
proba_valid = best_model_tree.predict_proba(features_valid_scale)
proba_one_valid = proba_valid[:, 1]
roc_auc = roc_auc_score(target_valid, proba_one_valid)

thresh_list.append(0.5)
f1_list.append(round(best_f1, 3))
accuracy_list.append(round(accur, 3))
roc_auc_list.append(round(roc_auc, 3))

# выведем результаты
print('max_depth: ', depth)
print('max_features: ', max_feature)
print('Accuracy: {:.3f}'.format(accur))
print('F1: {:.3f}'.format(best_f1))
print('ROC_AUC: {:.3f}'.format(roc_auc))
```

## Борьба с дисбалансом

В связи с небольшим количеством объектов в данных предлагается применить несколько методов для борьбы с дисбалансом:

- **простой апсемплинг** с дублированием объектов миноритарного класса;
- алгоритм **Smote**;
- подбор оптимального **порога** классификатора;
- **взвешивание классов** в логистической регрессии с использованием гиперпараметра **class_weight**.

Для сравнения двух подходов апсемплинга будут использованы те же модели что и на этапе **2.5**. Так как модели на основе лесов очень чувствительны к изменению в данных (в данном случае был применен апсемплинг) попробуем снова подобрать гиперпараметры для лучшей модели в тех же диапазонах значений.

### Простой Upsampling

```python
# запишем в переменные количество объектов каждого класса
count_class_0, count_class_1 = target_train.value_counts()
# разобъем выборку на отдельные классы
features_0 = features_train_scale[target_train==0]
features_1 = features_train_scale[target_train==1]
target_0 = target_train[target_train==0]
target_1 = target_train[target_train==1]
# увеличим количество объектов класса 1 дублированием для features и target
features_upsampled = pd.concat([features_0, features_1.sample(count_class_0, replace=True, random_state=rand_st)])
target_upsampled = pd.concat([target_0, target_1.sample(count_class_0, replace=True, random_state=rand_st)])
# перемешаем данные 
features_train_upsampled, target_train_upsampled = shuffle(features_upsampled, target_upsampled, random_state=rand_st)

print('Размер обучающей выборки после апсемплинга: ', features_train_upsampled.shape, target_train_upsampled.shape) 
print('Баланс классов в обучающей выборке: \n', target_train_upsampled.value_counts())
```

#### Модель DecissionTree

```python
%%time 

# инициализируем перемнную для лучшей оценки f1-меры 
best_f1 = 0 
                      
# подберем лучший гиперпараметр max_depth в цикле 
for max_feature_ in range(1,13):
    for depth_ in range(1,12):
        model_tree_up = DecisionTreeClassifier(max_depth=depth_, max_features=max_feature_, random_state=rand_st)
        model_tree_up.fit(features_train_upsampled, target_train_upsampled)
        predictions_valid = model_tree_up.predict(features_valid_scale)
      
        f1 = f1_score(target_valid, predictions_valid)
        
        # запишем в переменные гиперпараметры лучшей модели и саму модель
        if f1 > best_f1:
            best_f1 = f1
            max_feature = max_feature_
            depth = depth_
            best_model_tree_up = model_tree_up
        
# запишем в переменные и итоговые списки метрики лучшей модели 
predictions_valid = best_model_tree_up.predict(features_valid_scale)
accur = accuracy_score(target_valid, predictions_valid)
# для метрики ROC-AUC подсчитаем вероятности отнесения к классу
proba_valid = best_model_tree_up.predict_proba(features_valid_scale)
proba_one_valid = proba_valid[:, 1]
roc_auc = roc_auc_score(target_valid, proba_one_valid)

thresh_list.append(0.5)
f1_list.append(round(best_f1, 3))
accuracy_list.append(round(accur, 3))
roc_auc_list.append(round(roc_auc, 3))

# выведем результаты
print('max_depth: ', depth)
print('max_features: ', max_feature)
print('Accuracy: {:.3f}'.format(accur))
print('F1: {:.3f}'.format(best_f1))
print('ROC_AUC: {:.3f}'.format(roc_auc))
```

Так же, как метод борьбы с дисбалансом классов, применим **метод подбора оптимального порога** классификации. Для наглядности будут построены графики зависимости метрик от выбранного порога. 

```python
# построим график зависимости метрик от порога классификатора 
# так как график будет построен и для дркугих моделей напишем функцию отрисовки графика и подбора порога
def metric_tresh_graf(model): 
    scores_data = pd.DataFrame()
    for threshold in np.arange(0, 1, 0.01):
        proba_valid = model.predict_proba(features_valid_scale)
        proba_one_valid = proba_valid[:, 1] 
        predict_valid = proba_one_valid > threshold
        
        accur = accuracy_score(target_valid, predict_valid)
        roc_auc = roc_auc_score(target_valid, proba_one_valid)
        f1 = f1_score(target_valid, predict_valid) 
    
        tmp_scr_data = pd.DataFrame({'thresh': [threshold], 'accur': [accur], 'roc_auc': [roc_auc], 'f1': [f1]})
        scores_data = scores_data.append(tmp_scr_data)
        best_thresh = scores_data.nlargest(1, columns=['f1'])['thresh']
        
        #print("Threshold = {:.2f} | F1 = {:.3f}, ROC-AUC = {:.3f}, Accuracy = {:.3f}".format(threshold, f1, roc_auc, accur))
    
    plt.figure(figsize=(12, 8))
    plt.plot(scores_data['thresh'], scores_data['accur'], label='Accuracy')
    plt.plot(scores_data['thresh'], scores_data['roc_auc'], label='ROC-AUC')
    plt.plot(scores_data['thresh'], scores_data['f1'], label='F1')
    plt.axvline(x=best_thresh[0], linestyle='--', lw=1, color='black')
    plt.axhline(y=max(scores_data['f1']), linestyle='--', lw=1, color='black')
    plt.legend(loc="upper right")
    plt.ylim(0,1)
    plt.xlim(0,1)
    plt.xlabel("Порог")
    plt.ylabel("Значение метрики (качество)")
    plt.title("График зависимости метрик от порога классификатора DecisionTreeClassifier")
    plt.show()
    
    # запишем в итоговые списки метрики c оптимальным порогом 
    thresh_list.append(best_thresh[0])
    f1_list.append(round(max(scores_data['f1']), 3))
    accuracy_list.append(round(scores_data.nlargest(1, columns=['f1'])['accur'][0], 3))
    roc_auc_list.append(round(scores_data.nlargest(1, columns=['f1'])['roc_auc'][0], 3))
    
    # отобразим значения порога и метрик при наибольшем значении метрики F1
    display(scores_data.nlargest(1, columns=['f1']))
```

```python
# вызовем функцию отрисовки графика и подбора порога для лучшей модели DecisionTree
metric_tresh_graf(best_model_tree_up)
```

Из графика видно, что оптимальный порог при котором достигается максимальное качество модели по метрике **F1** составляет **0.58** . При этом метрика **Accuracy** = **0.815** и **ROC-AUC** = **0.838**. Таким образом путем подбора порога было достигнуто не только увеличение **F1 меры** до **0.619** (`на 0.017`) по сравнению с базовым порогом в **0.5**, но и метрики **Accuracy**.

### Smote

Алгоритм **Smote** так же относится к методам **апсемплинга** (увеличения) объектов **наименьшего класса**, но использует информацию о ближайших объектах этого же класса (**миноритарного**) для генерации новых объектов поблизости. 

```python
smote = SMOTE(sampling_strategy='minority', random_state=rand_st)
x_sm, y_sm = smote.fit_resample(features_train_scale, target_train)

print('Размер обучающей выборки после апсемплинга алгоритмом Smote: ', x_sm.shape, y_sm.shape) 
print('Баланс классов в обучающей выборке: \n', y_sm.value_counts())
```

#### Модель DecissionTree

```python
%%time 

# инициализируем перемнную для лучшей оценки f1-меры 
best_f1 = 0 
                      
# подберем лучший гиперпараметр max_depth в цикле 
for max_feature_ in range(1,13):
    for depth_ in range(1,12):
        model_tree_sm = DecisionTreeClassifier(max_depth=depth_, max_features=max_feature_, random_state=rand_st)
        model_tree_sm.fit(x_sm, y_sm)
        predictions_valid = model_tree_sm.predict(features_valid_scale)
      
        f1 = f1_score(target_valid, predictions_valid)
        
        # запишем в переменные гиперпараметры лучшей модели и саму модель
        if f1 > best_f1:
            best_f1 = f1
            max_feature = max_feature_
            depth = depth_
            best_model_tree_sm = model_tree_sm
        
# запишем в переменные и итоговые списки метрики лучшей модели 
predictions_valid = best_model_tree_sm.predict(features_valid_scale)
accur = accuracy_score(target_valid, predictions_valid)
# для метрики ROC-AUC подсчитаем вероятности отнесения к классу
proba_valid = best_model_tree_sm.predict_proba(features_valid_scale)
proba_one_valid = proba_valid[:, 1]
roc_auc = roc_auc_score(target_valid, proba_one_valid)

thresh_list.append(0.5)
f1_list.append(round(best_f1, 3))
accuracy_list.append(round(accur, 3))
roc_auc_list.append(round(roc_auc, 3))

# выведем результаты
print('max_depth: ', depth)
print('max_features: ', max_feature)
print('Accuracy: {:.3f}'.format(accur))
print('F1: {:.3f}'.format(best_f1))
print('ROC_AUC: {:.3f}'.format(roc_auc))
```

```python
metric_tresh_graf(best_model_tree_sm)
```

#### Модель RandomForest

```python
%%time

# инициализируем перемнную для лучшей оценки f1 меры
best_f1 = 0

# подберем лучшие гиперпараметры в цикле в определенном диапазоне значений
for depth_ in range(1, 11):
    for est_ in range(1, 35):
        model_rand_forest_sm = RandomForestClassifier(random_state=rand_st, max_depth=depth_, \
                                                   n_estimators=est_, max_features=8, n_jobs=-1)
        model_rand_forest_sm.fit(x_sm, y_sm)
        predictions_valid = model_rand_forest_sm.predict(features_valid_scale)
        f1 = f1_score(target_valid, predictions_valid)
        
        # запишем в переменные гиперпараметры лучшей моделиа и саму модель
        if f1 > best_f1:
            best_f1 = f1
            depth = depth_
            est = est_
            best_model_rand_forest_sm = model_rand_forest_sm

# запишем в переменные и итоговые списки метрики лучшей модели 
predictions_valid = best_model_rand_forest_sm.predict(features_valid_scale)
accur = accuracy_score(target_valid, predictions_valid)
# для метрики ROC-AUC подсчитаем вероятности отнесения к классу
proba_valid = best_model_rand_forest_sm.predict_proba(features_valid_scale)
proba_one_valid = proba_valid[:, 1]
roc_auc = roc_auc_score(target_valid, proba_one_valid)

thresh_list.append(0.5)
f1_list.append(round(best_f1, 3))
accuracy_list.append(round(accur, 3))
roc_auc_list.append(round(roc_auc, 3))

# выведем на экран результаты
print('max_depth: ', depth)
print('estimators: ', est)
print('Accuracy: {:.3f}'.format(accur))
print('F1: {:.3f}'.format(best_f1))
print('ROC_AUC: {:.3f}'.format(roc_auc))
```
